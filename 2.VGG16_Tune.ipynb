# 3.Preprocess

Preprocess the image following requirements of VGG-16

```python
train_ds_u = train_ds.unbatch()

lst_x = []
lst_y = []
for batch_images, batch_labels in train_ds_u :
  A = tf.keras.applications.vgg16.preprocess_input(batch_images)
  B = batch_labels
  lst_x.append(A)
  lst_y.append(B)

x_train = tf.stack(lst_x)
y_train = tf.stack(lst_y)
```

# 4.Load the pretrained model (only the feature extractor)

```python
# https://keras.io/api/applications/vgg/
img_w,img_h = 224,224 
vgg_extractor = tf.keras.applications.vgg16.VGG16(weights = "imagenet", include_top=False,input_shape = (img_w, img_h,3)) 

vgg_extractor.summary()
```

### 4.1 Tuning Feature Extraction layer

```python
# All layer = False
vgg_extractor.trainable = False

# Let's take a look to see how many layers are in the base model
print("Number of layers in the base model: ", len(vgg_extractor.layers))

# Fine-tune from this layer onwards
start_from = 10
for layer in vgg_extractor.layers[start_from:]:
  layer.trainable = True
```

```python
# recheck layer.trainable

for i,layer in enumerate(vgg_extractor.layers):  
    print( f"Layer {i}: name = {layer.name} , trainable = {layer.trainable}" )
```

### 4.2 Data Augment

```python
dataaug = tf.keras.Sequential([
    tf.keras.layers.RandomFlip(mode='horizontal_and_vertical', name='rand_flip'),
    tf.keras.layers.RandomTranslation(height_factor=0.2, width_factor=0.2, fill_mode='constant', name='rand_trans'),
    #tf.keras.layers.RandomRotation(factor=0.2, fill_mode='constant', name='rand_rot')
    ], name='dataaug' )
```

```python
dataaug
```

```python
input_1channel = (502, 502, 1)       # 1-channel image ภาพขาวดำ
input_3channel = (224, 224, 3)       # 3-channel image ภาพสี

input_dim = input_3channel            # choose the target dimension for this example
output_dim = 4                     # the number of output classes (single-label multi-class classification)
```

```python
# Input layer
inputs = tf.keras.Input(shape=input_dim)

# Data augment
x = dataaug(inputs)

# Feature Extraction
x = vgg_extractor(x)
x = tf.keras.layers.Conv2D( filters=32, kernel_size=(4,4), strides=(1,1), padding="same", activation='relu', name='conv1' )(x)
x = tf.keras.layers.Conv2D( filters=64, kernel_size=(4,4), strides=(1,1), padding="same", activation='relu', name='conv2' )(x)
x = tf.keras.layers.Conv2D( filters=32, kernel_size=(4,4), strides=(1,1), padding="same", activation='tanh', name='conv3' )(x)

# Flatten layer
x = tf.keras.layers.Flatten()(x)

# Classifier
x = tf.keras.layers.Dense(1024, activation="relu")(x)
x = tf.keras.layers.Dense(1024, activation="tanh")(x)
x = tf.keras.layers.Dropout(0.1)(x)

# Output layer
outputs = tf.keras.layers.Dense(output_dim, activation="softmax")(x)

# Construct the main model เอามาประกอบร่าง # output ตัวใหม่ที่เรา set ค่าไว้
model = tf.keras.models.Model(inputs=inputs, outputs=outputs)
```

```python
model.summary()
```

# 5.Compile and train the model

```python
initial_learning_rate = 0.0001 # Defaults to 0.001

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=initial_learning_rate),
    loss="sparse_categorical_crossentropy",
    metrics=["acc"]
    )
```

```python
checkpoint_filepath = "bestmodel_epoch{epoch:02d}_valloss{val_loss:.2f}.hdf5"
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_filepath,
                                                               save_weights_only=False, 
                                                               monitor='val_acc', 
                                                               mode='max',
                                                               save_best_only=True)
```

```python
%%time
history = model.fit(
    x = x_train,
    y = y_train,
    batch_size = batch_size,
    epochs=15, verbose=1,
    validation_data=[x_val, y_val],
    callbacks=[model_checkpoint_callback]
    )
```

```python
# Summarize history for accuracy
plt.figure(figsize=(15,5))
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Train accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.grid()
plt.show()

# Summarize history for loss
plt.figure(figsize=(15,5))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Train loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper right')
plt.grid()
plt.show()
```
